## Read 'AAPL' stock prices

```python
import pandas as pd

aapl_stock_data=pd.read_csv("stock_data_aapl_2014_2016.csv",index_col=0)
# Convert index to datetime
aapl_stock_data.index=pd.to_datetime(aapl_stock_data.index)
aapl_stock_data.head()
```

## Read sentiment score of 'AAPL' 

```python
apple_daily_sentiment_score=pd.read_csv('apple_daily_sentiment.csv',index_col=0)
#convert index to datetime
apple_daily_sentiment_score.index=pd.to_datetime(apple_daily_sentiment_score.index)

apple_daily_sentiment_score.head()
```

```python
# Join the two dataframes: aapl_stock_data and apple_daily_sentiment_score
prices=pd.concat([aapl_stock_data,apple_daily_sentiment_score],axis=1)
prices
```

```python
prices.dropna(inplace=True)
prices.head()
```

## Generate trading signals

```python
# Create a signal column which would hold the buy/sell signals
prices['signal']=0


```

```python
prices
```

```python
# Generate a buy signal whenever the sentiment score is more than 0.33
prices.loc[prices.score>=0.25,'signal']=1
# Generate a sell signal whenever the sentiment score is less than 0.33
prices.loc[prices.score<=-0.25,'signal']=-1

prices.head(5)

```

## Calculate returns of the strategy

```python
# Calculate 'AAPL' returns
prices['return']=prices['Open'].pct_change()
```

```python
prices.head(10)
```

```python
# Calculate strategy returns
prices['strategy_return']=prices['signal'].shift(1)*prices['return']
prices
```

```python
prices=prices.dropna()
prices.head(2)
```

### Plot the strategy returns

```python
# Import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

# Plot the strategy returns
(prices['strategy_return']+1).cumprod().plot(figsize=(10, 6))

# Label the graph
plt.title('Strategy Returns')
plt.grid()
plt.xlabel('Date')
plt.ylabel('Percentage Returns')
plt.show()
```

### Sharpe Ratio Calculation

```python
#Import numpy
import numpy as np

#The annual risk-free rate of 2% is assumed
risk_free_rate=0.02

#total number of trading days in a year is 252
trading_days=252

# Calculate the daily risk-free return by dividing the annual risk-free return by number of trading days
daily_risk_free_return = risk_free_rate/trading_days

# Calculate the excess returns by subtracting the daily returns by daily risk-free return
excess_daily_returns = prices['strategy_return'] - daily_risk_free_return

# Calculate the sharpe ratio using the given formula
# It is multiplied by square root of trading days to get the annualized value
sharpe_ratio = (excess_daily_returns.mean() /
                excess_daily_returns.std()) * np.sqrt(trading_days)
```

```python
print('The Sharpe ratio is %.2f' % sharpe_ratio)
```
